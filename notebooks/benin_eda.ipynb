Outlier Detection & Basic Cleaning

```python
df.describe()
```

```python
cols = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

for col in cols:
    plt.figure(figsize=(6,4))
    df.boxplot(column=col)
    plt.title(f"Box Plot of {col}")
    plt.ylabel(col)
    plt.show()
```

```python
cols_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

z_scores = np.abs(stats.zscore(df[cols_to_check], nan_policy='omit'))

# Flag rows where any column has |Z| > 3
outliers = (z_scores > 3).any(axis=1)


print(f"Number of outlier rows: {outliers.sum()}")
df_outliers = df[outliers]
df_outliers.head()
```

```python
# Drop rows flagged as outliers
df_clean = df[~outliers].copy()  # keep rows where outliers == False

# Check new shape
print("Original rows:", df.shape[0])
print("Rows after dropping outliers:", df_clean.shape[0])
```

```python
df_cleaned= df_clean[(df_clean["GHI"] >= 0) & (df_clean["DNI"] >= 0) & (df_clean["DHI"] >= 0)]
print("Rows after dropping negative solar irradiance values :", df_cleaned.shape[0])
```

Time Series Analysis

```python
df_cleaned['Timestamp'] = pd.to_datetime(df_cleaned['Timestamp'])
df_cleaned = df_cleaned.sort_values('Timestamp')
```

```python
unique_count = df_cleaned['Timestamp'].nunique()
print(f"Number of unique timestamps: {unique_count}")
```

```python
start_date = df_cleaned['Timestamp'].min()
end_date = df_cleaned['Timestamp'].max()

print(f"Date range: {start_date} to {end_date}")
```

```python
df_cleaned['Month'] = df_cleaned['Timestamp'].dt.month
df_cleaned['Hour'] = df_cleaned['Timestamp'].dt.hour
```

```python
# Daily average
daily_avg = df_cleaned.resample('D', on='Timestamp')[['GHI','DNI','DHI','Tamb']].mean()

# Monthly average
monthly_avg = df_cleaned.groupby('Month')[['GHI','DNI','DHI','Tamb']].mean()
```

```python
plt.figure(figsize=(14,6))
plt.plot(daily_avg.index, daily_avg['GHI'], label='GHI')
plt.plot(daily_avg.index, daily_avg['DNI'], label='DNI')
plt.plot(daily_avg.index, daily_avg['DHI'], label='DHI')
plt.plot(daily_avg.index, daily_avg['Tamb'], label='Tamb (C)')

plt.title('Daily Average Solar Irradiance and Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(df_cleaned['Timestamp'], df_cleaned['GHI'], label='GHI', alpha=0.8)
plt.plot(df_cleaned['Timestamp'], df_cleaned['DNI'], label='DNI', alpha=0.8)
plt.plot(df_cleaned['Timestamp'], df_cleaned['DHI'], label='DHI', alpha=0.8)
plt.plot(df_cleaned['Timestamp'], df_cleaned['Tamb'], label='Tamb (C)', alpha=0.8)
plt.title('Solar Irradiance and Temperature over Time')
plt.xlabel('Timestamp')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_cleaned['Timestamp'] = pd.to_datetime(df_cleaned['Timestamp'])
df_cleaned['month'] = df_cleaned['Timestamp'].dt.month
df_cleaned['hour'] = df_cleaned['Timestamp'].dt.hour
df_cleaned['day'] = df_cleaned['Timestamp'].dt.day
```

```python
monthly_avg = df_cleaned.groupby('month')[['GHI','DNI','DHI','Tamb']].mean()

monthly_avg.plot(kind='bar', figsize=(10,6))
plt.title('Monthly Average Solar and Temperature Patterns')
plt.xlabel('Month')
plt.ylabel('Average Value')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
df_cleaned['GHI_roll'] = df_cleaned['GHI'].rolling(window=60).mean()  # e.g., 60-min average

plt.figure(figsize=(14,5))
plt.plot(df_cleaned['Timestamp'], df_cleaned['GHI'], label='GHI (Raw)', alpha=0.4)
plt.plot(df_cleaned['Timestamp'], df_cleaned['GHI_roll'], label='GHI (Rolling Mean)', color='red')
plt.title('GHI Over Time with Rolling Average')
plt.legend()
plt.show()
```

```python
hourly_avg = df_cleaned.groupby('hour')[['GHI','DNI','DHI','Tamb']].mean()

hourly_avg.plot(kind='bar', figsize=(12,6))
plt.title('Average Solar and Temperature Patterns by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Value')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Cleaning Impact

```python
mod_clean = df_cleaned.groupby('Cleaning')[['ModA', 'ModB']].mean()
print(mod_clean)
```

```python
mod_clean.plot(kind='bar', figsize=(6,5), color=['#69b3a2', '#ffcc66'])
plt.title('Average ModA & ModB Readings Before and After Cleaning')
plt.xlabel('Cleaning Flag (0 = Before, 1 = After)')
plt.ylabel('Average Sensor Reading')
plt.xticks(rotation=0)
plt.legend(title='Module')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Correlation & Relationship Analysis

Wind & Distribution Analysis

Wind rose

```python
wind_data = df_cleaned.dropna(subset=['WS', 'WD'])

plt.figure(figsize=(7,7))
ax = WindroseAxes.from_ax()
ax.bar(wind_data['WD'], wind_data['WS'], normed=True, opening=0.8, edgecolor='white')
ax.set_title('Wind Rose: Direction and Speed Distribution', fontsize=12)
ax.set_legend(title='Wind Speed (m/s)')
plt.show()
```

Histogram

```python
plt.figure(figsize=(12,5))

# Histogram for GHI
plt.subplot(1,2,1)
sns.histplot(df_cleaned['GHI'], bins=30, kde=True, color='gold')
plt.title('Distribution of Global Horizontal Irradiance (GHI)')
plt.xlabel('GHI')
plt.ylabel('Frequency')

# Histogram for WS
plt.subplot(1,2,2)
sns.histplot(df_cleaned['WS'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Wind Speed (WS)')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Bubble Chart

```python
plt.figure(figsize=(10,6))

plt.scatter(
    x=df_cleaned['GHI'], 
    y=df_cleaned['Tamb'], 
    s=df_cleaned['RH'],        # bubble size = RH
    alpha=0.6,
    c=df_cleaned['RH'],        # color by RH for extra info
    cmap='viridis',
    edgecolors='w'
)

plt.title('GHI vs Temperature (Bubble size = RH)')
plt.xlabel('GHI (W/m)')
plt.ylabel('Tamb (C)')
plt.colorbar(label='Relative Humidity (%)')
plt.grid(True)
plt.show()
```
